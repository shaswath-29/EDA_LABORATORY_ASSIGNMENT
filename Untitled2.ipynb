```python
import pandas as pd
import numpy as np

df1cp = pd.DataFrame({'Jersey number': [7, 10, 18, 45, 12, 34], 'Rankings': [1, 34, 23, 4, 5, 11]})
df2cp = pd.DataFrame({'Jersey number': [22, 35, 67, 49, 12, 33], 'Rankings': [32, 67, 864, 34, 21, 22]})

df1fp = pd.DataFrame({'Jersey number': [7, 10, 21, 37, 99], 'Rankings': [39, 80, 42, 176, 78]})
df2fp = pd.DataFrame({'Jersey number': [23, 49, 72, 64, 19], 'Rankings': [27, 47, 89, 15, 31]})
```

```python
dfcp = pd.concat([df1cp, df2cp], ignore_index=True)
dffp = pd.concat([df1fp, df2fp], ignore_index=True)

df = pd.concat([dfcp, dffp], axis=1)
df
```

```python
dfcp = pd.concat([df1cp, df2cp], ignore_index=True)
dffp = pd.concat([df1fp, df2fp], ignore_index=True)

df = dfcp.merge(dffp, how='outer')
df
```

```python
dfcp = pd.concat([df1cp, df2cp], ignore_index=True)
dffp = pd.concat([df1fp, df2fp], ignore_index=True)

df = dfcp.merge(dffp, how='left')
df
```

```python
import pandas as pd
import numpy as np

financial_data = np.arange(15).reshape((3, 5))
metrics = ['Revenue', 'Net Income', 'Free Cash Flow']
companies = ['Apple', 'Microsoft', 'Google', 'Amazon', 'Meta']

dframe1 = pd.DataFrame(financial_data, index=metrics, columns=companies)
print(dframe1)

stacked = dframe1.stack()
print(stacked)

unstacked = stacked.unstack()
print(unstacked)
```

```python
standard_delivery = pd.Series([101, 102, 103, 104], index=['Alpha', 'Beta', 'Gamma', 'Delta'])
express_delivery = pd.Series([201, 202, 203], index=['Epsilon', 'Zeta', 'Eta'])

frame2 = pd.concat([standard_delivery, express_delivery], keys=['Standard_Tier', 'Express_Tier'])
print(frame2)
print(frame2.unstack())
```
